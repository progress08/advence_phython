```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
students = {
    'Name': ['John', 'Emma', 'Mike', 'Sara', 'Tom'],
    'Score': [75, 88, 92, 67, 81]
}
df = pd.DataFrame(students)
df
```

### Task 6 - Line Plot

```python
plt.plot(df['Name'], df['Score'], marker='o')
plt.xlabel('Student')
plt.ylabel('Score')
plt.title('Student Scores')
plt.show()
```

### Task 7 - Bar Chart

```python
plt.bar(df['Name'], df['Score'])
plt.xlabel('Student')
plt.ylabel('Score')
plt.title('Student Scores')
plt.show()
```
